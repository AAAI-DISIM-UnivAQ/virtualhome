```python

import sys
import glob
import json
```

```python
files = glob.glob('../programs_processed_precond_nograb_morepreconds/initstate/*/*.json')
```

```python
sys.path.append('../')
import script_utils
```

```python
all_loc = []
location = []
for file in files:
    loc = []
    with open('{}'.format(file), 'r') as f:
        initstate = json.load(f)
    for conds in initstate:
        if list(conds)[0] == 'location':
            loc.append(conds['location'][1])
    loc = [tuple(x) for x in loc]
    location += [x[0] for x in loc]
    loc_list = loc
    loc = list(set(loc))
    all_loc.append((len(loc), len(loc_list)))
location = list(set(location))
print(location)
```

```python
cont = 0
for it, file in enumerate(files):
    elem = file.replace('initstate', 'withconds').replace('.json', '.txt')
    nconds = all_loc[it]
    if nconds[0] > 1:
        with open(elem, 'r') as f:
            content = f.readlines()
        numconds = int(content[0])
        prog = content[numconds+5:]
        objects = []
        for instr in prog:
            _, obj, idi = script_utils.parseStrBlock(instr.strip())
            
            for elem in zip(obj, idi):
                if elem[0] not in location:
                    objects.append(elem)
        objects = list(set(objects))
        if len(objects) > nconds[1]:
            print(file)
            print(objects, nconds)
            print(''.join(content))
            print('\n')
            cont += 1
print(cont)
```
