```python
import json
graph_dict = json.load(open('example_graphs/TrimmedTestScene6_graph.json'))
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import networkx as nx

G = nx.DiGraph()
```

```python
for node in graph_dict["nodes"]:
    G.add_node(node["id"], category=node["category"], class_name=node["class_name"])
```

```python
edge_type2color = {
    "INSIDE": "red", 
    "ON": "blue", 
    "FACING": "green", 
    "CLOSE": "yellow", 
}
```

```python

for edge in graph_dict["edges"]:
    edge_type = edge["relation_type"]
    G.add_edge(edge["from_id"], edge["to_id"], color=edge_type2color[edge_type] if edge_type in edge_type2color else "black")
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw_random(G, node_size=20, arrows=True, width=0.1)
```
